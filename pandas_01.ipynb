# Intro to Pandas

When working with tabular data, such as data stored in spreadsheets or databases, pandas is the right tool for you. pandas will help you to explore, clean, and process your data. In pandas, a data table is called a `DataFrame`.

```python
# creating a Dataframe
df1 = pd.DataFrame(data = np.arange(0, 20).reshape(5, 4), index = ["Row1", "Row2", "Row3", "Row4", "Row5"], columns =["Column1", "Column2", "Column3", "Column4"])
```

# Retriving the elements

we can retrive the elements using 3 methods


1.   using comlumn_index(slicing technique)
2.   using row_index(`loc` method)
3.   using row_column_index(`iloc` method)





```python
df1.head()
```

```python
df1["Column2"]
```

```python
type(df1["Column2"])
```

```python
# when you are fetching more than one colum then use nested suared brackets

df1[["Column1", "Column2", "Column3"]]
```

```python
type(df1[['Column1', 'Column2', 'Column3']])
```

```python
# retriving the rows using index_name method (loc[])
# we can get row and column as well but it need string indexers not numbric indexers
df1.loc["Row1"]
```

```python
type(df1.loc["Row1"])
```

```python
df1.loc[["Row1", "Row2"]]
```

```python
type(df1.loc[["Row1", "Row2"]])
```

```python
df1.loc[["Row1", "Row2", "Row3"], ["Column1", "Column2"]]
```

```python
# Retriving the data using index and column number (.iloc method)

df1.iloc[0:3, 1:4]
```

```python
df1.head()
```

```python
df1.loc[["Row2", "Row3"], ["Column2", "Column3"]]
```

```python
# iloc requires numeric indexers

df1.iloc[1: 3, 1: 3]
```

```python
df1.iloc[:,:]
```

```python
type(df1.iloc[1: 3, 1: 3])
```

```python
df1.iloc[0,]
```

```python
type(df1.iloc[0,])
```

```python
df1.iloc[:, :]
```

```python
type(df1.iloc[: , :])
```

```python
df1.iloc[:,0::3]
```
